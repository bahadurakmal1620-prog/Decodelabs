```python
# Cell 1: Imports and Dataset Creation

import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Our job roles dataset (this is your raw_skills.csv equivalent)
data = {
    'job_role': [
        'Data Scientist',
        'Machine Learning Engineer',
        'Backend Developer',
        'Frontend Developer',
        'DevOps Engineer',
        'Cloud Architect',
        'Data Analyst',
        'Cybersecurity Analyst',
        'Mobile App Developer',
        'AI Research Scientist'
    ],
    'skills': [
        'python sql machine_learning data_analysis statistics numpy pandas tensorflow',
        'python tensorflow pytorch deep_learning algorithms neural_networks numpy',
        'python java sql apis rest django databases nodejs',
        'javascript html css react nodejs ui_ux web_design',
        'aws docker kubernetes linux ci_cd automation git cloud',
        'aws azure cloud networking infrastructure automation devops',
        'sql excel data_visualization power_bi statistics reporting python',
        'networking linux firewalls ethical_hacking cryptography security',
        'java kotlin swift flutter mobile ios android ui_ux',
        'python research deep_learning mathematics optimization neural_networks tensorflow'
    ]
}

df = pd.DataFrame(data)
print("✅ Dataset Created Successfully!")
print(df)
```

```python
# Cell 2: TF-IDF Vectorization

# Initialize the TF-IDF Vectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the skills column
tfidf_matrix = vectorizer.fit_transform(df['skills'])

# See the vocabulary (all unique skills found)
vocab = vectorizer.get_feature_names_out()

print("✅ TF-IDF Matrix Built Successfully!")
print(f"\n📐 Matrix Shape: {tfidf_matrix.shape}")
print(f"   → {tfidf_matrix.shape[0]} job roles  x  {tfidf_matrix.shape[1]} unique skills")
print(f"\n📚 Vocabulary (all unique skills): \n{vocab}")
```

```python
# Cell 3: User Input & Cosine Similarity Scoring

# ✏️ User enters their 3 skills here
user_skills = input("Enter your 3 skills separated by spaces (e.g. python sql machine_learning): ")

# Convert user input into a TF-IDF vector (same vocabulary space)
user_vector = vectorizer.transform([user_skills])

# Calculate cosine similarity between user vector and all job role vectors
similarity_scores = cosine_similarity(user_vector, tfidf_matrix)

# Flatten the scores into a simple array
scores = similarity_scores.flatten()

print("\n✅ Similarity Scores Calculated!")
print("\n📊 Raw Scores for each Job Role:")
for i, score in enumerate(scores):
    print(f"   {df['job_role'][i]:<30} → Score: {score:.4f}")
```

```python
# Cell 4: Sorting, Filtering & Final Output

TOP_N = 3  # You can change this to 5 if you want more results

# Step 3: Sort scores in descending order
ranked_indices = np.argsort(scores)[::-1]

# Step 4: Filter - take only Top N
top_indices = ranked_indices[:TOP_N]

print("=" * 50)
print("🎯 TOP {} CAREER RECOMMENDATIONS FOR YOU".format(TOP_N))
print("=" * 50)
print(f"📥 Your Skills: {user_skills}\n")

for rank, idx in enumerate(top_indices, start=1):
    score = scores[idx]
    role = df['job_role'][idx]
    skills = df['skills'][idx]
    match_percent = round(score * 100, 2)

    print(f"🏆 Rank {rank}: {role}")
    print(f"   Match Score  : {score:.4f}")
    print(f"   Match %      : {match_percent}%")
    print(f"   Role Skills  : {skills}")
    print("-" * 50)

print("\n✅ Recommendation Engine Complete!")
```

```python
# Cell 5: Full Reusable Recommendation Function

def recommend_career(user_skills_input, top_n=3):
    print("=" * 55)
    print("🤖 TECH STACK CAREER RECOMMENDER - DecodeLabs")
    print("=" * 55)

    # Step 1: Ingestion - vectorize user input
    user_vector = vectorizer.transform([user_skills_input])

    # Step 2: Scoring - cosine similarity
    scores = cosine_similarity(user_vector, tfidf_matrix).flatten()

    # Check if all scores are zero (Cold Start problem)
    if scores.max() == 0:
        print("⚠️  No matches found! Your skills are not in our vocabulary.")
        print("💡 Try using these skills:")
        print(", ".join(vectorizer.get_feature_names_out()))
        return

    # Step 3: Sorting
    ranked_indices = np.argsort(scores)[::-1]

    # Step 4: Filtering - Top N only
    top_indices = ranked_indices[:top_n]

    print(f"\n📥 Input Skills : {user_skills_input}")
    print(f"🎯 Top {top_n} Recommended Career Paths:\n")

    for rank, idx in enumerate(top_indices, start=1):
        score = scores[idx]
        role = df['job_role'][idx]
        match_percent = round(score * 100, 2)

        print(f"  🏆 Rank {rank}: {role}")
        print(f"     Match     : {match_percent}%")
        print(f"     {'█' * int(match_percent // 5):<20} {match_percent}%")
        print()

    print("=" * 55)
    print("✅ Powered by TF-IDF + Cosine Similarity")
    print("=" * 55)


# ▶️ Run the recommender
user_input = input("Enter your skills (space separated): ")
recommend_career(user_input, top_n=3)
```
