```python
# The Knowledge Base — your chatbot's "brain"
responses = {
    "hello": "Hey there! How can I help you?",
    "hi": "Hi! What's up?",
    "how are you": "I'm just a bot, but I'm running perfectly!",
    "what is ai": "AI stands for Artificial Intelligence — machines that simulate human decision-making.",
    "who made you": "I was built by an AI Engineer in training at DecodeLabs!",
    "bye": "Goodbye! See you next time.",
    "help": "I can answer questions about AI, greet you, and have a basic chat!"
}

print("✅ Knowledge base loaded with", len(responses), "intents.")
```

```python
# The Sanitizer — cleans raw user input
def sanitize(raw_input):
    clean = raw_input.lower().strip()
    return clean

# Test it
test_inputs = ["HELLO", "  Hi  ", "What is AI?", "BYE"]

for test in test_inputs:
    print(f"Raw: '{test}' → Clean: '{sanitize(test)}'")
```

```python
# Enhanced sanitizer — also removes punctuation
def sanitize(raw_input):
    clean = raw_input.lower().strip()
    clean = clean.replace("?", "").replace("!", "").replace(".", "")
    return clean

# The Response Engine using .get() — O(1) lookup + fallback in one shot
def get_response(raw_input):
    clean = sanitize(raw_input)
    reply = responses.get(clean, "I don't understand that yet. Try asking something else!")
    return reply

# Test it
test_inputs = ["HELLO", "  Hi  ", "What is AI?", "BYE", "random gibberish"]

for test in test_inputs:
    print(f"You: {test}")
    print(f"Bot: {get_response(test)}")
    print()
```

```python
# The Complete Rule-Based AI Chatbot
print("🤖 DecodeLabs Chatbot is online!")
print("   Type 'exit' to shut down.\n")

while True:
    raw_input = input("You: ")

    # Exit strategy — clean break command
    if sanitize(raw_input) == "exit":
        print("Bot: Shutting down... Goodbye! 👋")
        break

    # Get and print response
    reply = get_response(raw_input)
    print(f"Bot: {reply}\n")
```
